```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Load the data
Daniel = pd.read_csv('/content/kenya_accidents.csv')
# Extract the independent variable (X) and dependent variable (y)
X = Daniel[['AGE']]
y = Daniel['NO OF VICTIMS']
# Create and fit a linear regression model
regressor = LinearRegression()
regressor.fit(X, y)
# Plot the data points
Daniel.plot(x='AGE', y='NO OF VICTIMS', style='o', label='Data Points')
# Overlay the regression line
plt.plot(X, regressor.predict(X), color='red', linewidth=2, label='Regression Line')
plt.xlabel('AGE')
plt.ylabel('NO OF VICTIMS')
plt.legend()
plt.show()
```
